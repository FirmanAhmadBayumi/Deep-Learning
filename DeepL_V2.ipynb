```python
# Confusion Matrix values from the new image
confusion_matrix_new = {
    'license_plate': {'TP': 461, 'FN': 8, 'FP': 2},
    'cars': {'TP': 258, 'FN': 66, 'FP': 106},
    'motorcycle': {'TP': 88, 'FN': 11, 'FP': 10},
    'truck': {'TP': 5, 'FN': 4, 'FP': 8}
}

# Calculate Total Instances for TN calculation
total_instances_new = sum([confusion_matrix_new[key]['TP'] + confusion_matrix_new[key]['FN'] for key in confusion_matrix_new])

# Function to calculate metrics
def calculate_metrics(TP, FN, FP, total):
    TN = total - (TP + FN + FP)
    accuracy = (TP + TN) / total
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0
    f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return {
        'accuracy': accuracy,
        'recall': recall,
        'precision': precision,
        'specificity': specificity,
        'f1_score': f1_score
    }

# Calculating metrics for each class
metrics_new = {}
for key, values in confusion_matrix_new.items():
    metrics_new[key] = calculate_metrics(values['TP'], values['FN'], values['FP'], total_instances_new)

# Display the calculated metrics
for class_name, metric in metrics_new.items():
    print(f"\nMetrics for {class_name.capitalize()}:")
    print(f"Accuracy: {metric['accuracy']:.2%}")
    print(f"Recall: {metric['recall']:.2%}")
    print(f"Precision: {metric['precision']:.2%}")
    print(f"Specificity: {metric['specificity']:.2%}")
    print(f"F1-Score: {metric['f1_score']:.2%}")
```

```python
import os
import shutil

# Path ke dataset
base_path = "/content/yolov10/vehicle-and-license-plate-3"

# Buat folder baru untuk menyimpan dataset yang akan disatukan
dataset_images_path = "/content/yolov10/dataset/images"
dataset_labels_path = "/content/yolov10/dataset/labels"
os.makedirs(dataset_images_path, exist_ok=True)
os.makedirs(dataset_labels_path, exist_ok=True)

# Fungsi untuk memindahkan file dengan validasi apakah file ada
def copy_files_from_split(split):
    images_dir = os.path.join(base_path, split, "images")
    labels_dir = os.path.join(base_path, split, "labels")

    if os.path.exists(images_dir):
        image_files = os.listdir(images_dir)
        for img_file in image_files:
            # Pindahkan file gambar
            src_img_path = os.path.join(images_dir, img_file)
            if os.path.exists(src_img_path):
                shutil.copy(src_img_path, dataset_images_path)
            else:
                print(f"Gambar tidak ditemukan: {src_img_path}")

            # Pindahkan file label
            label_file = img_file.replace(".jpg", ".txt")
            src_label_path = os.path.join(labels_dir, label_file)
            if os.path.exists(src_label_path):
                shutil.copy(src_label_path, dataset_labels_path)
            else:
                print(f"Label tidak ditemukan: {src_label_path}")

# Pindahkan file dari train, val, test
copy_files_from_split("train")
copy_files_from_split("valid")
copy_files_from_split("test")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Path ke dataset asli
base_path = "/content/yolov10/vehicle-and-license-plate-3"

# Buat folder baru untuk menyimpan dataset yang sudah di-split ulang
new_base_path = "/content/yolov10/dataset"
os.makedirs(os.path.join(new_base_path, "train/images"), exist_ok=True)
os.makedirs(os.path.join(new_base_path, "train/labels"), exist_ok=True)
os.makedirs(os.path.join(new_base_path, "val/images"), exist_ok=True)
os.makedirs(os.path.join(new_base_path, "val/labels"), exist_ok=True)
os.makedirs(os.path.join(new_base_path, "test/images"), exist_ok=True)
os.makedirs(os.path.join(new_base_path, "test/labels"), exist_ok=True)

# Kumpulkan semua gambar dari train, valid, dan test
def collect_images(split):
    images_dir = os.path.join(base_path, split, "images")
    return [os.path.join(images_dir, img) for img in os.listdir(images_dir)]

all_images = collect_images("train") + collect_images("valid") + collect_images("test")

# Lakukan split menjadi 80% train dan 20% (val + test)
train_imgs, temp_imgs = train_test_split(all_images, test_size=0.2, random_state=42)

# Split 20% sisanya menjadi 10% validasi dan 10% test
val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.5, random_state=42)

# Fungsi untuk memindahkan file gambar dan label ke folder yang sesuai
def move_files(file_list, dest_split):
    for img_file in file_list:
        # Dapatkan nama file
        img_filename = os.path.basename(img_file)

        # Pindahkan gambar
        shutil.copy(img_file, os.path.join(new_base_path, dest_split, "images", img_filename))

        # Pindahkan label yang sesuai
        label_filename = img_filename.replace(".jpg", ".txt")
        label_file = img_file.replace("images", "labels").replace(".jpg", ".txt")
        shutil.copy(label_file, os.path.join(new_base_path, dest_split, "labels", label_filename))

# Pindahkan file yang sudah di-split ke folder yang sesuai
move_files(train_imgs, "train")
move_files(val_imgs, "val")
move_files(test_imgs, "test")

print("Dataset berhasil di-split menjadi 8:1:1 dan dipindahkan ke folder yang sesuai.")
```
